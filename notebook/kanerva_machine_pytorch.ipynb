```python
import numpy as np
import os, sys
import glob

import torch
from torch.autograd import Variable
import torch.nn.functional as F
import torchvision
from torchvision import transforms
import torchvision.datasets as dset
import torch.optim as optim
from torch import nn
from torch.distributions import MultivariateNormal

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# for loading and batching MNIST dataset
def setup_data_loaders(batch_size=128, use_cuda=False):
    root = './data'
    download = True
    trans = transforms.ToTensor()
    train_set = dset.MNIST(root=root, train=True, transform=trans,
                           download=download)
    test_set = dset.MNIST(root=root, train=False, transform=trans)

    kwargs = {'num_workers': 1, 'pin_memory': use_cuda}
    train_loader = torch.utils.data.DataLoader(dataset=train_set,
        batch_size=batch_size, shuffle=True, **kwargs)
    test_loader = torch.utils.data.DataLoader(dataset=test_set,
        batch_size=batch_size, shuffle=False, **kwargs)
    return train_loader, test_loader
```

```python
def kronecker_product(t1, t2):
    """
    Computes the Kronecker product between two tensors.
    See https://en.wikipedia.org/wiki/Kronecker_product
    """
    t1_height, t1_width = t1.size()
    t2_height, t2_width = t2.size()
    out_height = t1_height * t2_height
    out_width = t1_width * t2_width

    tiled_t2 = t2.repeat(t1_height, t1_width)
    expanded_t1 = (
        t1.unsqueeze(2)
          .unsqueeze(3)
          .repeat(1, t2_height, t2_width, 1)
          .view(out_height, out_width)
    )

    return expanded_t1 * tiled_t2
```

```python
class Encoder1(nn.Module):
    '''
    return q(y|x) and 2C-encoded vec
    '''
    def __init__(self, D_in, H, C):
        super(Encoder1, self).__init__()
        self.linear1 = torch.nn.Linear(D_in, H)
        self.linear2y = torch.nn.Linear(H, 2*C)
        self.linear2y_mean = torch.nn.Linear(2*C, C)
        self.linear2y_var = torch.nn.Linear(2*C, C)
        self.linear2z = torch.nn.Linear(H, C)

    def forward(self, x):
        x = F.relu(self.linear1(x))
        y = F.relu(self.linear2y(x))
        z = F.relu(self.linear2z(x))
        y_mean = self.linear2y_mean(y)
        y_var = self.linear2y_var(y)
        eps = Variable(torch.randn(y_var.size()[0], y_var.size()[1]))
        _y = y_mean + torch.exp(y_var / 2)*eps # Reparametazation Trick
        return _y, z, y_mean, y_var

class Encoder2(nn.Module):
    '''
    return q(z|x, (y, M))
    '''
    def __init__(self, C, H):
        super(Encoder2, self).__init__()
        self.linear1 = torch.nn.Linear(2*C, H)
        self.linear2 = torch.nn.Linear(H, C)
    
    def forward(self, x):
        x = F.relu(self.linear1(x))
        return F.relu(self.linear2(x))

class Decoder(nn.Module):
    '''
    return q(x|z)
    '''
    def __init__(self, D_in, H, D_out):
        super(Decoder, self).__init__()
        self.linear1 = torch.nn.Linear(D_in, H)
        self.linear2 = torch.nn.Linear(H, 300)
        self.linear3 = torch.nn.Linear(300, D_out)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = F.relu(self.linear1(x))
        return self.sigmoid(self.linear3(F.relu(self.linear2(x))))
```

```python
class Memory(nn.Module):
    def __init__(self, n_address, n_memory_vec, n_address_vec, m_value = 1.0):
        '''
        
        '''
        super(Memory, self).__init__()
        self.y_dim = n_memory_vec
        self.z_dim = n_memory_vec
        self.n_address = n_address
        self.n_memory_vec = n_memory_vec
        self.n_address_vec = n_address_vec
        self.m_value = 1.0
        self.R = torch.rand(self.n_address, self.n_memory_vec)
        self.U = torch.eye(self.n_address)
        self.V = torch.eye(self.n_memory_vec)
        self.M = None
        self.calculate_M()
        
        self.A = nn.Linear(self.n_address_vec, self.n_address)
        self.fc_y1 = nn.Linear(self.y_dim, self.n_address_vec//2)
        self.fc_y2 = nn.Linear(self.n_address_vec//2, self.n_address_vec)
        
    
    def forward(self, _input, batch_size):
        _y = self.fc_y1(_input)
        bt = self.fc_y2(_y+0.08*torch.randn(_y.size()))
        wt = self.A(bt)
        _M = self.M.sample().view(batch_size, self.n_address, self.n_memory_vec)
        bM = []
        for i in range(batch_size):
            bM.append(_M)
        bM = torch.cat(bM, 0)
        _mean = torch.bmm(wt.view(batch_size, 1, -1), bM) * self.m_value
#         print(_mean.size())
        return _mean.view(-1, self.n_memory_vec)
    
    
    def calculate_M(self, R=None, U=None, V=None):
        if R is not None:
            self.R = R
        if U is not None:
            self.U = U
        if V is not None:
            self.V = V
#         print(self.R.view(-1, self.n_address*self.n_memory_vec).data[0].numpy().shape)
#         print(kronecker_product(self.V, self.U).size())
        self.M = MultivariateNormal(self.R.view(-1, self.n_address*self.n_memory_vec), covariance_matrix=kronecker_product(self.V, self.U))
    
    
    def write(self, y_list, Z, v_sigma):
        W = None
        len_t = len(y_list)
        for _y in y_list:
            _y = self.fc_y1(_y)
            bt = self.fc_y2(_y)
            wt = self.A(bt)
            if W is None:
                W = wt.view(1, self.n_address)
            else:
                W = torch.cat((W, wt), 0)
        dd = Z - torch.matmul(W, self.R)
        sigma_c  = torch.matmul(W, self.U)
        sigma_gusai = torch.eye(len_t)
        sigma_z = torch.matmul(torch.matmul(W, self.U), torch.t(W)) + sigma_gusai * (v_sigma*v_sigma)
        R = self.R + torch.matmul(torch.matmul(torch.t(sigma_c),torch.inverse(sigma_z)), dd)
        U = self.U - torch.matmul(torch.matmul(torch.t(sigma_c),torch.inverse(sigma_z)), sigma_c)
        self.calculate_M(R=R, U=U)
```

```python
class KanervaMachine(torch.nn.Module):
    '''
    '''
    def __init__(self, n_address, n_memory_vec, n_address_vec, n_input=784, n_hidden=100, batch_size=1):
        super(KanervaMachine, self).__init__()
        self.encoder1 = Encoder1(n_input, n_hidden, n_memory_vec)
        self.encoder2 = Encoder2(n_memory_vec, n_hidden)
        self.decoder = Decoder(D_in=n_memory_vec, H=n_hidden, D_out=n_input)
        self.memory = Memory(n_address, n_memory_vec, n_address_vec)
        
        self.n_address = n_address
        self.n_memory_vec = n_memory_vec
        self.n_address_vec = n_address_vec
        
        self.batch_size = batch_size

    def forward(self, x_input, len_n):
        _x = x_input
        total_loss = None
        y_kl_loss = None
        z_list = []
        for i in range(len_n):
            _y, z1, y_mean, y_var = self.encoder1(_x)
            z_mean = self.memory(_y, self.batch_size)
            eps = Variable(torch.randn(z_mean.size()[0], z_mean.size()[1]))
            z2 = z_mean + torch.exp(torch.ones(z_mean.size()[0], z_mean.size()[1])/2.)*eps # Reparametazation Trick
            _z = torch.cat((z1, z2), 1)
            _z = self.encoder2(_z)
            output = self.decoder(_z)
            _x = output
            if y_kl_loss is None:
                y_kl_loss = torch.mean(0.5 * torch.sum(torch.exp(y_var) + y_mean**2 - 1. - y_var, 1))
            else:
                y_kl_loss = y_kl_loss + torch.mean(0.5 * torch.sum(torch.exp(y_var) + y_mean**2 - 1. - y_var, 1))
            z_list.append(_z-z_mean)
        y_kl_loss = torch.mean(y_kl_loss)
        z_list = torch.cat(z_list, 0)
        z_list_mean = torch.mean(z_list, 0).view(-1, len_n)
        z_list_var = torch.var(z_list, 0).view(-1, len_n)
        z_kl_loss = torch.mean(0.5 * torch.sum(torch.exp(z_list_var) - 1. - z_list_var, 1))
        return _x, y_kl_loss, z_kl_loss
    
    def write(self, x_seq):
        len_t = x_seq.size()[0]
        len_input = x_seq.size()[1]
        y_list = []
        z_list = []
        for i in range(len_t):
            _y, z1, y_mean, y_var = self.encoder1(x_seq[i].view(1, len_input))
            _z = torch.cat((z1, z1), 1)
            _z = self.encoder2(_z)
            z_list.append(_z)
            y_list.append(_y)
        Z = torch.cat(z_list, 0)
        self.memory.write(y_list, Z, 1)
    
    def init_memory(self):
        self.memory.calculate_M(R=torch.rand(self.n_address, self.n_memory_vec), U=torch.eye(self.n_address))
```

# Training

```python
# Run options
LEARNING_RATE = 5.0e-4
USE_CUDA = False
NB_ITERATION = 100000
BATCH_SIZE = 1
T_LENGTH = 10
N_LENGTH = 10
INPUT_SIZE = 784
```

```python
km = KanervaMachine(n_address=20, n_memory_vec=20, n_address_vec=10, n_input=784, n_hidden=10)
train_loader, test_loader = setup_data_loaders(batch_size=T_LENGTH, use_cuda=USE_CUDA)
solver = optim.Adam(km.parameters(), lr=LEARNING_RATE)

recon_loss = nn.BCELoss()
# recon_loss.size_average = False
```

```python
i = 0
km.init_memory()
for idx, (x, l) in enumerate(train_loader):
    input_x_seq = x.view(-1, INPUT_SIZE)
    km.write(input_x_seq)
    if idx > 200:
        break
for idx, (x, l) in enumerate(train_loader):
    input_x_seq = x.view(-1, INPUT_SIZE)
    total_x_loss = None
    total_y_loss = None
    total_z_loss = None
    solver.zero_grad()
    for jdx in range(T_LENGTH):
        input_x = input_x_seq[jdx].view(1, INPUT_SIZE)
        output_x, y_loss, z_loss = km(input_x, N_LENGTH)
        x_loss = recon_loss(output_x, input_x)
        if total_x_loss is None:
            total_x_loss = x_loss
        else:
            total_x_loss = total_x_loss + x_loss
        if total_y_loss is None:
            total_y_loss = y_loss
        else:
            total_y_loss = total_y_loss + y_loss
        if total_z_loss is None:
            total_z_loss = z_loss
        else:
            total_z_loss = total_z_loss + z_loss
    total_loss = (total_x_loss + total_y_loss + total_z_loss) / float(T_LENGTH)
#         total_loss = (total_x_loss*0.1 + total_y_loss*0.1 + total_z_loss)
    # Backward
    total_loss.backward()
    # Update
    solver.step()
    print(idx, total_x_loss, total_y_loss, total_z_loss)
    if idx % 100 == 0:
        output_x = output_x.view(28, 28).data.numpy()
        plt.figure()
        plt.imshow(output_x, cmap='Greys_r')
        plt.show()
```
